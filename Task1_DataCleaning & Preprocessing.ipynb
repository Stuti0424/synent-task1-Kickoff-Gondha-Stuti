```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Summer Internship\titanic.csv")
```

```python
df.head(10)
```

```python
print("\nDataset Shape:")
print(df.shape)
```

```python
print("\nColumn Names:")
print(df.columns)
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
columns = ['Age', 'Income', 'CCAvg']

for col in columns:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())
df
```

```python
columns = ['Embarked']

for col in columns:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])
df
```

```python
if 'Cabin' in df.columns:
    df.drop(columns=['Cabin'], inplace=True)
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
categorical_columns = ['Survived', 'Pclass', 'Sex']

for col in categorical_columns:
    if col in df.columns:
        df[col] = df[col].astype('category')
```

```python
print(df.dtypes)
```

```python
df.rename(columns={
    'PassengerId': 'Passenger_ID',
    'Pclass': 'Passenger_Class',
    'Name': 'Passenger_Name',
    'Sex': 'Gender'
}, inplace=True)
df
```

```python
print("\nCleaned Dataset Info:")
print(df.info())
```

```python
print("\nFirst 10 Rows After Cleaning:")
print(df.head(10))
```

```python
df.to_csv('cleaned_titanic_dataset.csv', index=False)

print("\nCleaned dataset is saved")
```
